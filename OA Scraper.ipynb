```python
!pip install telethon

import os
import shutil
import asyncio
from telethon import TelegramClient

# Provide your Telegram API credentials
api_id = ''  # Replace with your actual API ID
api_hash = ''  # Replace with your actual API Hash
channel_username = ""  # Replace with the channel username

# Google Colab-specific output directory
output_folder = "/content/sorted_telegram_images"
os.makedirs(output_folder, exist_ok=True)

# List of company names (strict matching)
companies = [
    "Google", "Apple", "Tower Research", "Uber", "Directi", "Media.net", "Zeta", "Flock",
    "LinkedIn", "Microsoft", "Amazon", "Adobe", "Cloudera", "Twitter", "Flipkart", "Yahoo",
    "Rubrik", "Salesforce", "Slack", "Oracle", "MindTickle", "Paypal", "Rippling", "Goldman Sachs",
    "Intuit", "De Shaw", "Arcesium", "Walmart", "ServiceNow", "Xilinx", "AMD", "GitHub", "Nutanix",
    "InMobi", "NortonLifeLock", "Codenation", "Cure.Fit", "Intel", "Atlassian", "Qualcomm", "Visa",
    "eBay", "SumoLogic", "Expedia", "Paytm", "Swiggy", "Grab", "Morgan Stanley", "VMware", "NVIDIA",
    "DropBox", "HackerRank", "Urban Company", "Citicorp", "OYO", "Cisco", "Hotstar", "Hike messenger",
    "Ola", "MakeMyTrip", "Samsung", "Times Internet", "Zomato", "BYJU’S", "Dream11", "Siemens",
    "Mentor Graphics", "Junglee Games", "Dunzo", "Rivigo", "Arista Networks", "Airtel", "NetApp", "SAP",
    "Synopsys", "GreyOrange", "Unacademy", "Myntra", "ThoughtWorks", "Cadence", "CRED", "American Express",
    "Juniper Networks", "BrowserStack", "Citrix", "RedHat", "Ixigo", "Grofers", "Snapdeal", "ClearTax",
    "BankBazaar", "Livspace", "1mg", "Master Card", "BigBasket", "PayU", "Sirion Labs", "AJio", "Practo",
    "Info Edge", "JpMorgan Chase", "Mobikwik", "Proptiger", "Akamai", "ClearTrip", "Dell", "BharatPe",
    "Zerodha", "Groww", "Car Dekho", "Zoho", "BlackBuck", "Yatra", "Goibibo", "Accolite", "HP", "Philips",
    "Delhivery", "GE", "UnitedHealth Group", "Palo Alto Networks", "[24]7.ai", "Innovaccer", "Lenskart",
    "RazorPay", "MPL", "Dailyhunt", "Meesho", "FirstCry", "ShareChat", "Vedantu"
]

# Convert list to lowercase for case-insensitive matching
companies_lower = {company.lower(): company for company in companies}

def get_company_name(text):
    """Finds company name in text using exact match (case-insensitive)."""
    text_lower = text.lower()
    for key in companies_lower:
        if key in text_lower:
            return companies_lower[key]
    return "Unknown"

async def scrape_telegram():
    """Scrapes Telegram messages and sorts images into folders."""
    async with TelegramClient("session_name", api_id, api_hash) as client:
        async for message in client.iter_messages(channel_username, reverse=True):  # Scrape in reverse order
            if message.photo:
                file_path = await message.download_media(output_folder)
                company = get_company_name(message.text) if message.text else "Unknown"

                # Move image to respective company folder
                company_folder = os.path.join(output_folder, company)
                os.makedirs(company_folder, exist_ok=True)
                shutil.move(file_path, os.path.join(company_folder, os.path.basename(file_path)))

                print(f"Saved {file_path} → {company_folder}/")

# Run the async function in Colab
await scrape_telegram()

print("✅ Sorting complete!")
```

```python
import os
import shutil

# Update this path with your actual Google Drive location
base_dir = "/content/drive/My Drive/sorted_telegram_images"
unknown_dir = os.path.join(base_dir, "Unknown")

# New companies set
new_companies = {
    "Amex", "ZScaler", "TTL", "Barclay", "Twilio", "Citadel", "Mercari",
    "Amdocs", "VYMO", "Nomura", "AWS", "UBS", "Sony", "Expedia", "NPCI",
    "Qualcomm", "Wipro", "Cognizant", "TCS", "Infosys", "Accenture", "Deloitte"
}

# Create new folders if they don't exist
for company in new_companies:
    company_folder = os.path.join(base_dir, company)
    os.makedirs(company_folder, exist_ok=True)

# Process each image in "Unknown"
for img_file in os.listdir(unknown_dir):
    img_path = os.path.join(unknown_dir, img_file)

    # Check if the filename contains a company name
    matched_company = None
    for company in new_companies:
        if company.lower() in img_file.lower():  # Case-insensitive check
            matched_company = company
            break

    # Move file to matched company folder
    if matched_company:
        dest_path = os.path.join(base_dir, matched_company, img_file)
        shutil.move(img_path, dest_path)
        print(f"Moved {img_file} → {matched_company}")

print("Sorting complete.")
```
